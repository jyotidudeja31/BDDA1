## Experiment 1
Number of sample points required for a statistically significant result rises very fast as number of dimensions increase

```python
import numpy as np
```

```python
# 1. Data dimensions: a, b, c, d (4 dimensions)
a=np.arange(1,101)
b=np.arange(1,101)
c=np.arange(1,101)
d=np.arange(1,101)
```

```python
# 2. Get 1% of one dimensional space
#      Sample 1% of 'a' ie if 'a' has 100 points,
#        just, 1 data-point, is needed out of 100 points
#         to make a statistically significant decision
```

```python
# 3. Get 1% of two-dimensional space
space2 = np.max(a) * np.max(b)
space2
0.01 * space2  # equals (10% of a) X (10% of b)
               # That is for 2D space one needs
               #  10% of data-points from each dimension
               #   Or 100 points are needed to make
               #     statistically significant decision
```

```python
# 4. Get 1% of three dimensioanl space
space3 = np.max(a) * np.max(b) * np.max(c)
space3
0.01 * space3 # Greater than 0.2 * max(a) * 0.2* max(b) * 0.2 * max(c)
              # That is data-points required are 20% of dimension
              # If each dimension has a range of 100 possible values,
              #   then number of samples required will be
              #     20 * 20 * 20 = 8000
              #      Or 8000 points are needed to make
              #        statistically significant decision
```

```python
# 4. For a 10 dimensioanl space requirement is 63% of each dimension
# 5. And for a 100 dimensional space, it is 95% of each dim's range

#################################################################
```

## Experiment 2
For large dimensions all data is concentrated in the corners of hypersphers and hence  equidistant from center.

```python
#  Last amended on 20th June, 2021
#  My folder: C:\Users\ashokharnal\OneDrive\Documents\Curse_of_dimensionality
#  Illustrate curse of dimensionality
#  Related file: curse.R
#
#  Refer: http://www.edupristine.com/blog/curse-dimensionality
#  Objective:
#      For large dimensions all data is concentrated in the corners of space
#       Show that:
#        Hypersphere volume reduces drastically in relation to hypercube volume
#          as number of dimensions increase. Thus, most of the volume is in the
#           corners of hypercube so also the number of points. Center of hypercube
#            is empty.
# Ref: https://en.wikipedia.org/wiki/Volume_of_an_n-ball
```

image.png

image.png

```python
#  We draw sphere inside of cube & calculate
#   ratio of space occupied by sphere inside cube
```

```python
# Assume any dimension, r, of (half) hypercube side.
#   Within it we will inscribe a hypersphere of radius r
#     Thus sphere diameter: 2r (= one side of hypercube)
#      Any inscribed sphere will touch all sides of hypercube
#       leaving out corners
```

```python
import numpy as np
r = 1
pi = np.pi
```

```python
## 1
# 1.1 Dimension of cube: 2
#     'Volume' of two dimensional ball radius r
#     pi * r ** 2

hyper2 = pi * r ** 2
```

```python
# 1.2 Area of square of side 2r : (2r) ** 2
cube2 =  (2*r) ** 2
```

```python
# 1.3 Ratio of volume of sphere/volume of container
sphereVolToContainerVol2 = hyper2/cube2
sphereVolToContainerVol2  # Sphere occupies 78.5% of square
                          # So 78.5% of points are within sphere
```

```python
# 2. Dimension 3
# 2.1 Volume of three dimensional ball radius r (sphere)
#     (4/3)pi * r ** 3
hyper3 = (4/3)*pi * r ** 3
```

```python
cube3 =  (2*r) ** 3
# 2.2 Ratio of volume of sphere/volume of container
sphereVolToContainerVol3 = hyper3/cube3
sphereVolToContainerVol3  # So only 52.35% of points are within sphere
```

```python
# 3. Dimension 4
# 3.1 Volume of four dimensional ball radius r (sphere)
#     (pi ** 2) * r ** 4
hyper4 = (pi ** 2)/2 * (r ** 4)
```

```python
# 3.2 Area of hypercube of side 2r : (2r) ** 4
cube4 =  (2*r) ** 4
# 3.3 Ratio of volume of sphere/volume of container
sphereVolToContainerVol4 = hyper4/cube4
sphereVolToContainerVol4    # So only 30% of points are within sphere
```

```python

# 4. Dimension 5
# 4.1 Volume of four dimensional ball radius r (sphere)
#     (8* (pi ** 2)/15) * r ** 5
hyper5 = (8* (pi ** 2)/15) * r ** 5
```

```python
# 4.2 Area of hypercube of side 2r : (2r) ** 5
cube5 =  (2*r) ** 5
# 4.3 Ratio of volume of sphere/volume of container
sphereVolToContainerVol5 = hyper5/cube5
sphereVolToContainerVol5
```

```python
# 5. Dimension 6
# 5.1 Volume of four dimensional ball radius r (sphere)
#     (pi ** 3)/6 * r ** 6
hyper6 = (pi ** 3)/6 * r ** 6
```

```python
# 5.2 Area of hypercube of side 2r : (2r) ** 5
cube6 =  (2*r) ** 6
# 5.3 Ratio of volume of sphere/volume of container
sphereVolToContainerVol6 = hyper6/cube6
sphereVolToContainerVol6
```

```python
# 6. Plot the results
import matplotlib.pyplot as plt

# 6.1 List of ratios
ratios=[sphereVolToContainerVol2,
        sphereVolToContainerVol3,
        sphereVolToContainerVol4,
        sphereVolToContainerVol5,
        sphereVolToContainerVol6]

# 6.2 List of dimensions
x = range(2,7)

# 6.3 Plot dimensions against ratio
plt.plot(x,ratios)
plt.xlabel("No of dimensions of hypercube")
plt.ylabel("Hypersphere vol to hypercube vol ratio")
plt.show()
```
